```python
import pandas as pd
import plotly.express as px

# 1. Preparação dos Dados com Pandas (Copiado da célula anterior para garantir que df seja definido)
data = {
    "Fruta": ["Maçãs", "Laranjas", "Bananas", "Maçãs", "Laranjas", "Bananas"],
    "Quantidade": [4, 1, 2, 2, 4, 5],
    "Cidade": ["São Paulo", "São Paulo", "São Paulo", "Rio de Janeiro", "Rio de Janeiro", "Rio de Janeiro"]
}
df = pd.DataFrame(data)

# Ordenando os dados para melhor visualização (conforme boas práticas)
df = df.sort_values(by="Quantidade", ascending=False)

# 2. Criação do Gráfico com Plotly Express
fig = px.bar(
    df,
    x="Fruta",
    y="Quantidade",
    color="Cidade",
    barmode="group",
    title="Consumo de Frutas por Cidade",
    labels={"Quantidade": "Qtd Consumida", "Fruta": "Tipo de Fruta"},
    template="plotly_white"
)
fig.show()
```
